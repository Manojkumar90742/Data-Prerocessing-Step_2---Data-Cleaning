***Handling Missing Values***

***If our dataset have four independent variable and one dependent variable all features are have missing values.We must handling the missing values.***

```python
# If we have missing values in the target column/dependent variable.We must drop the record.

df=df.dropna(subset=['y'])
```

```python
# If we have missing values in the independent variable.We must handle the data.
# Replace the missing values in the categorical column by using mode values. mode value means repeated values.

cols=['x3']
df[cols]=df[cols].fillna(df.mode().iloc[0])
```

```python
# Replace the missing values in the categorical column by using mean,median,mode values.

cols1=['x1','x2','x4']
df[cols1]=df[cols1].fillna(df.mean())
```
